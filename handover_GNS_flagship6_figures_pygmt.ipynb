```python
import pygmt
import geopandas as gpd
import pandas as pd
import rasterio as rio
import xarray
import rioxarray
from pathlib import Path
import matplotlib.pyplot as plt
import contextily as ctx
```

```python
# plot exposure 
p_exposure = Path(r"/Volumes/GoogleDrive/My Drive/Workfolder/Work_tech/flaship6_gns/Exposure/Exposure_Napier_2020/WGS84")

buildings = gpd.read_file(p_exposure/ "Exposure_Napier_2020_point_wgs84.shp")
buildings.set_crs(epsg=4326, inplace=True)

# collect extent region
bound = buildings.total_bounds
REGION = [bound[0], bound[2], bound[1], bound[3]]
filename = "Exposure_Napier_2020_polygon_wgs84.gmt"
```

```python
fig = pygmt.Figure()
fig.basemap(region=REGION, projection="M20c", frame=True)
fig.grdimage("@earth_relief_01s", cmap='geo', shading=True)
fig.plot(data=str(p_exposure/filename), color="black", label="Exposure_Napier")
fig.legend(position="jTL+o0.2c", box='+gwhite+p1p') #JTR = j"inside", TR"TopRight"
fig.colorbar(frame=["x+lElevation", "y+lm"])
fig.show()# method="external"
```
